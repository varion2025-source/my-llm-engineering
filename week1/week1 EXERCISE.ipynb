```python
link_system_prompt = """
You are an expert in coding.
You will provided with a question.
You are able to explain the code and provide possible resources to futher study
"""
```

```python
def get_user_prompt(question_code):
    user_prompt = f"""
Here is the code question {question_code} -
Please provide a exaplaination
"""   
    return user_prompt
```

```python
def answer_code_question(question_code):
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": link_system_prompt},
            {"role": "user", "content": get_user_prompt(question_code)}
        ],       
    )
    result = response.choices[0].message.content    
    return result
    
```
